```python
import pandas as pd

df = pd.read_csv('../Data/Train_Data/2019_Draft.csv')
```

```python
df.head()
```

```python
df = df.rename(columns={"Drafted (tm/rnd/yr)": "Drafted"})
```

```python
draft = df.dropna(subset='Drafted')
draft.head()
```

```python
drafted = draft.dropna()
drafted
```

```python
drafted["round"] = df["Drafted"].str.extract("/ (\d+)", expand=False)
```

```python
drafted
```

```python
drafted = drafted.drop(columns=['College','Player-additional'], axis=1)
drafted
```

```python
drafted.to_csv(
    "2019_Drafted.csv",
    index=False,
    header=True,
    sep=",",
    na_rep="NA",
    quotechar='"',
    encoding="utf-8",
)
```

```python
df_av= pd.read_csv('../Data/Train_Data/2019_NFL_AV.csv')
df_av.head()
```

```python
full_df = drafted.merge(df_av, on='Player', how='left')
full_df.head()
```

```python
full_df.to_csv(
    "2019_players.csv",
    index=False,
    header=True,
    sep=",",
    na_rep="NA",
    quotechar='"',
    encoding="utf-8",
)
```
